```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
import os
import time
import datetime as dt
```

```python
# Creates lists for data to be stored 
date = []
place = []
mag = []
cdi = []
mmi = []
coordinates = []
lat = []
long = []

#Output data 
output_file = "../Data/earthquake.csv"

#importing the API call 
def earthquake(e):
    params = {"starttime": "1800-01-01", 
              "minmagnitude": "4.0", "maxradiuskm":600, "latitude": 34.21, "longitude":-118.57}
    data = requests.get(e, params = params)
    data = json.loads(data.text)
    return data

e = r"https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson"
a = earthquake(e)

#printing the properties of the data we need for analysis 
for i in (a["features"]):
    print(i["properties"]["time"], i["properties"]["place"],
    i["properties"]["cdi"], i["properties"]["mag"],i["properties"]["mmi"],i["geometry"]["coordinates"][0],i["geometry"]["coordinates"][1])
```

```python
#appending the data to the empty lists 
for i in (a["features"]):
    date.append(i["properties"]["time"])
    place.append(i["properties"]["place"])
    cdi.append(i["properties"]["cdi"])
    mag.append(i["properties"]["mag"])
    mmi.append(i["properties"]["mmi"])
    long.append(i["geometry"]["coordinates"][0])
    lat.append(i["geometry"]["coordinates"][1])
    
# create a dictionary for establishing dataframe
earthquake_data = {
    "City":place,
    "Date":date,
    "Magnitude": mag,
    "MMI": mmi,
    "CDI":cdi,
    "Latitude":lat,
    "Longitude":long
}

earthquake_df = pd.DataFrame(earthquake_data)

earthquake_df
```

```python
#converting the date format 
date_list = earthquake_df['Date'].tolist()
converted_list = []

for i in date_list:
    converted = time.strftime('%Y-%m-%d %H:%M:%S', time.gmtime(i/1000.0))
    converted_list.append(converted)
```

```python
earthquake_df["Time"] = ""

earthquake_df["Time"] = converted_list
earthquake_df.head()
```

```python
earthquake_df.rename(columns={'Date': 'Original Date','Time': 'Date'}, inplace=True)

earthquake_df
```

```python
#splitting the dates 
earthquake_df[['Dates','Time']] = earthquake_df.Date.str.split(" ",expand=True,)

#creating columns for year, month and day 
earthquake_df[['Year','Month','Day']] = earthquake_df.Dates.str.split("-",expand=True,)

#dropping columns to clean the dataframe
final_earthquake_df = earthquake_df.drop(['Original Date','Date', 'Dates'], axis=1)

#sorting the columns in ascending order
final_earthquake_df = final_earthquake_df.sort_values(by=['Year','Month','Day'], ascending=[True, True,True])
```

```python
#Exports DataFrame as a csv file
final_earthquake_df.to_csv('Resources/earthquakes.csv', index= False)

final_earthquake_df
```

```python
# Fort Tejon API Call
# Clear lists for data to be stored 
date = []
place = []
mag = []
cdi = []
mmi = []
coordinates = []
lat = []
long = []

# 1857 Fort Tejon Earthquake Data & 30 Years Prior, Within 150km
def earthquake(e):
    params = {"starttime": "1817-01-01", "endtime": "1857-01-20",
              "minmagnitude": "4.0", "maxradiuskm":150, "latitude": 35.700, "longitude":-120.30}
    data = requests.get(e, params = params)
    data = json.loads(data.text)
    return data

e = r"https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson"
a = earthquake(e)

for i in (a["features"]):
    print(i["properties"]["time"], i["properties"]["place"],
    i["properties"]["cdi"], i["properties"]["mag"],i["properties"]["mmi"],i["geometry"]["coordinates"][0],i["geometry"]["coordinates"][1])
```

```python
# Pre-dictionary storage loop
for i in (a["features"]):
    date.append(i["properties"]["time"])
    place.append(i["properties"]["place"])
    cdi.append(i["properties"]["cdi"])
    mag.append(i["properties"]["mag"])
    mmi.append(i["properties"]["mmi"])
    lat.append(i["geometry"]["coordinates"][1])
    long.append(i["geometry"]["coordinates"][0])
    
# Create a dictionary for establishing dataframe
fort_tejon = {
    "City":place,
    "Epoch Time":date,
    "Magnitude": mag,
    "MMI": mmi,
    "CDI":cdi,
    "Latitude":lat,
    "Longitude":long
}

fort_tejon_raw = pd.DataFrame(fort_tejon)
fort_tejon_raw.head()
```

```python
# Converting epoch time to human-readable format
epoch_list = fort_tejon_raw['Epoch Time'].tolist()
converted_list = []

for i in epoch_list:
    converted = time.strftime('%Y-%m-%d %H:%M:%S', time.gmtime(i/1000.0))
    converted_list.append(converted)
```

```python
# Adding human-readable time to DF
fort_tejon_raw['Date'] = converted_list
fort_tejon_raw
```

```python
# Preparing df for final export
fort_tejon_dirty = fort_tejon_raw
fort_tejon_dirty[['City', 'State']] = fort_tejon_raw.City.str.split(",",expand=True,)
fort_tejon_dirty[['Dates', 'Time']] = fort_tejon_raw.Date.str.split(" ",expand=True,)
fort_tejon_dirty[['Year', 'Month', 'Day']] = fort_tejon_raw.Dates.str.split("-",expand=True,)
fort_tejon_dirty['Earthquake'] = fort_tejon_dirty['City'] + ", " + fort_tejon_dirty['Year']
```

```python
# Selecting for relevant columns
fort_tejon_df = fort_tejon_dirty[['Earthquake', 'State', 'Magnitude',
                                     'Latitude', 'Longitude', 'Day', 'Month', 'Year', 'Time', 'Dates']]

# Final Export & Presentation
fort_tejon_df.to_csv('Resources/fort_tejon_1857.csv', index=False)
fort_tejon_df
```
